### Necessary libraries


```python
import numpy as np
import matplotlib.pyplot as plt 
import random 
import math
```

### Part 1: Making a large number of n = 24 paths to get an idea of probability

```python
# A variable to keep track of when 25 occurs
count_25 = 0

# Repeating the path generation enough times for statisitcal significance
for i in range(0,10000):

# Each path has a root value of 1
    path = [1]

# Flipping a coin n times and appending the value of the last entry of the path + toss value to the path
    for i in range(0,24):
        coin = [1,2]
        flip  = random.choice(coin)
        path.append(path[-1]+flip)

# Checking for 25 each time a path is generated
    if 25 in path:
        count_25 = count_25 + 1
        
    i  = i +1

# Returning how often 25 occurs
print(count_25)
```

### Part 2: Generating a plot for the probability of a number being in a path for the first 25 positive integers

```python
# array containing all the numbers 1 to 25
x = np.arange(1,26,1)
x = x.tolist()

y = []

# for each number, calculating its probability using the equation we found 
for i in range(0,25):
    y_add = 2/3 + (1/3)*(-0.5)**(i)
    y.append(y_add)

# standard plot of x against y
fig, ax = plt.subplots(figsize = (10,6))
plt.plot(x,y,color = '#21deaf')

# basic visusal elements
ax.grid(visible = True)
ax.spines[['right', 'top']].set_visible(False)
plt.xticks(np.arange(1,26,1))
plt.yticks(np.arange(0.4,1.1,0.1))
plt.title("Probability of x being in a path of length 25")
plt.xlabel("Number")
plt.ylabel("Probability")
```

### Part 3: Investigating how length of path impacts probability of 25 appearing

```python
# only need to do this for 12 to 25, and to make for loop easier, the trivial case is added to the array to start
x = np.arange(12,25,1)
x = x.tolist()
y = [1/2**12]

# using the counting method and summing with all previous values to get probability for each n
for i in range(1,13):
    y_new = math.comb(i + 12, 24 - i - 12)/2**(i+12)
    y_add = y_new + y[i-1]
    y.append(y_add)

# standard x/y plot of n against probability
fig, ax = plt.subplots(figsize = (10,6))
plt.plot(x,y,color = '#21deaf')

# various visual elements
ax.grid(visible = True)
ax.spines[['right', 'top']].set_visible(False)
plt.xticks(np.arange(12,25,1))
plt.yticks(np.arange(0,1.1,0.1))
plt.title("Probability of x being in a path of length <25")
plt.xlabel("Length of path (n)")
plt.ylabel("Probability")
```

### Extension 1: How adding in a 3-jump changes how often 25 occurs

```python
total_count = []

# repeating the experiment for all necessary n
for j in range(8,25):
    # A variable to keep track of when 25 occurs
    count_25 = 0
    
    # Repeating the path generation enough times for statisitcal significance
    for i in range(0,10000):
    
    # Each path has a root value of 1
        path = [1]
    
    # Flipping a coin n times and appending the value of the last entry of the path + toss value to the path
        for i in range(0,j):

    # making the coin '3 sided' but still fair
            coin = [1,2,3]
            flip  = random.choice(coin)
            path.append(path[-1]+flip)
    
    # Checking for 25 each time a path is generated
        if 25 in path:
            count_25 = count_25 + 1
            
        i  = i +1

    # adding the count for each n (normalised) to the array for graphing
    total_count.append(count_25/10000)
    j = j + 1

fig, ax = plt.subplots(figsize = (10,6))
plt.plot(np.arange(8,25,1),total_count,color = '#21deaf')
ax.grid(visible = True)
ax.spines[['right', 'top']].set_visible(False)
plt.xticks(np.arange(8,25,1))
plt.title("Hpw often 25 occurs for a path of length n")
plt.xlabel("Length of path (n)")
plt.ylabel("How often 25 appears (normalised)")
```

### Extension 2: How giving the coin a bias to 1 impacts how often 25 occurs

```python
total_count = []

# repeating the experiment for all necessary n
for j in range(12,25):
    # A variable to keep track of when 25 occurs
    count_25 = 0
    
    # Repeating the path generation enough times for statisitcal significance
    for i in range(0,10000):
    
    # Each path has a root value of 1
        path = [1]
    
    # Flipping a coin n times and appending the value of the last entry of the path + toss value to the path
        for i in range(0,j):

    # giving the coin a bias by adding an extra side with a 1 on
            coin = [1,1,2]
            flip  = random.choice(coin)
            path.append(path[-1]+flip)
    
    # Checking for 25 each time a path is generated
        if 25 in path:
            count_25 = count_25 + 1
            
        i  = i +1

    # adding the count for each n (normalised) to the array for graphing
    total_count.append(count_25/10000)
    j = j + 1

fig, ax = plt.subplots(figsize = (10,6))
plt.plot(np.arange(12,25,1),total_count,color = '#21deaf')
ax.grid(visible = True)
ax.spines[['right', 'top']].set_visible(False)
plt.xticks(np.arange(12,25,1))
plt.title("How often 25 occurs for a path of length n")
plt.xlabel("Length of path (n)")
plt.ylabel("How often 25 appears (normalised)")
```
